```python
import pandas as pd
import re

csv_file = 'markup.csv'
df = pd.read_csv(csv_file, sep='\t')

dataset = []
tokens = []
annotations = []
```

```python
#Функция проверки разметки для колонок KE_0 - KE_4
def check_consistency_and_quality(dataset):
    inconsistent_annotations = []
    missing_annotations = []

    for sentence in dataset:
        tokens = sentence['tokens']
        annotations = sentence['annotations']

        # Итерируемся через токены и аннотации
        for i, (token, annotation) in enumerate(zip(tokens, annotations)):
            # Проверяем на отсуствие аннотации или неправильную разметку
            if annotation == 'O':
                continue
            elif annotation.startswith('B-'):
                entity_type = annotation[2:]
                # Проверяем если следующий токен это I-tag с той же сущностью или O-tag, означающий что сущность закончилась
                if i == len(annotations) - 1 or (annotations[i + 1] != 'I-' + entity_type and annotations[i + 1] != 'O'):
                    missing_annotations.append((token, annotation))
            elif annotation.startswith('I-'):
                entity_type = annotation[2:]
                # Проверяем если предыдущий токен это B-tag или I-tag с той же сущностью  
                if i == 0 or (annotations[i - 1] != 'B-' + entity_type and annotations[i - 1] != 'I-' + entity_type):
                    inconsistent_annotations.append((token, annotation))
            else:
                print(f"Invalid annotation found: {annotation}")

    return inconsistent_annotations, missing_annotations
```
